### Target Variable Distribution

```python
value_counts_sorted = df['Hazard'].value_counts().sort_index()
print(value_counts_sorted)
```

Observation

* The target variable has a wide range of values, ranging from 1 to 69.
* The majority class is labeled as 1 with 15,229 occurrences, followed by classes 4 and 2 with 6,561 and 4,939 occurrences, respectively.
* Some classes have relatively low frequencies, such as classes 35, 51, 64, 33, which have only 1 occurrence each.
* There is a gradual decrease in the frequency of classes as the class label increases.
